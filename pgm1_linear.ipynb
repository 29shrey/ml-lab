```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1)
```

```python
myplt=plt.figure(figsize=(15,10))
ax1 = myplt.add_subplot(2,2,1)
ax1=plt.scatter(
 df['TV'],
 df['sales'],
 c='green' 
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()

myplt=plt.figure(figsize=(15,10))
ax2 = myplt.add_subplot(2,2,2)
ax2=plt.scatter(
 df['radio'],
 df['sales'],
 c='yellow'  
)
plt.xlabel("Money spent on Radio ads ($)")
plt.ylabel("Sales ($)")
plt.show()

myplt=plt.figure(figsize=(15,10))
ax3 = myplt.add_subplot(2,2,3)
ax3=plt.scatter(
 df['newspaper'],
 df['sales'],
 c='orange'  
)
plt.xlabel("Money spent on Newspaper ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
x = df['TV'].values.reshape(-1, 1)
y = df['sales'].values.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
reg = LinearRegression()
reg.fit(x_train, y_train)

print("Slope(m): ", reg.coef_[0][0])
print("Intercept(c): ", reg.intercept_[0])
print("Linear model of TV vs sales is: Y={:.3} + {:.2}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(x_test)
plt.figure(figsize=(15, 10))
plt.scatter(
 x_test,
 y_test,
 c='green'
)
plt.plot(
 x_test,
 predictions,
 c='blue',
 linewidth=2
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(y_test,predictions))
print("Root Mean Squared Error = ",rmse)
```
