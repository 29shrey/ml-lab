```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
x = df.drop(['Class'], axis = 1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
id3 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
model = id3.fit(x_train, y_train)
y_pred = id3.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, matthews_corrcoef
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(matthews_corrcoef(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True)
```

```python
cart = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
model = cart.fit(x_train, y_train)
y_pred1 = cart.predict(x_test)
```

```python
print(classification_report(y_test, y_pred1))
print(accuracy_score(y_test, y_pred1))
print(matthews_corrcoef(y_test, y_pred1))
sns.heatmap(confusion_matrix(y_test, y_pred1), annot = True)
```
