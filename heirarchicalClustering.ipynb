```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_data = pd.DataFrame(iris.data)
iris_data.columns = iris.feature_names
iris_data['Flower_type'] = iris.target
```

```python
x = iris_data.iloc[:, [0, 1, 2, 3]].values
y = iris_data.iloc[:, 4].values
```

```python
plt.scatter(x[y==0,0], x[y==1,0], s=100, c='red', label='Type1')
plt.scatter(x[y==1,0], x[y==1,1], s=100, c='yellow', label='Type2')
plt.scatter(x[y==2,0], x[y==2,1], s=100, c='blue', label='Type3')
plt.legend()
plt.show()
```

```python
sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('dendogram AGNES')
plt.xlabel('samples')
plt.ylabel('euclidean distance')
plt.show()
```

```python
sch.dendrogram(sch.linkage(x, method = 'complete'))
plt.title('dendogram DIANA')
plt.xlabel('samples')
plt.ylabel('euclidean distance')
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
cluster.fit(x)
labels = cluster.labels_
labels
```

```python
plt.scatter(x[labels==0,0], x[labels==0,1], s=100, color='red', label='Type1')
plt.scatter(x[labels==1,0], x[labels==1,1], s=100, color='blue', label='Type2')
plt.scatter(x[labels==2,0], x[labels==2,1], s=100, color='yellow', label='Type3')
plt.title('Clusters AGNES')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'complete')
cluster.fit(x)
labels = cluster.labels_
labels
```

```python
plt.scatter(x[labels==0,0], x[labels==0,1], s=100, color='red', label='Type1')
plt.scatter(x[labels==1,0], x[labels==1,1], s=100, color='blue', label='Type2')
plt.scatter(x[labels==2,0], x[labels==2,1], s=100, color='yellow', label='Type3')
plt.title('Clusters DIANA')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
